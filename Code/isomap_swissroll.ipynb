# Isomap Dimensionality Reduction on Swiss Roll Data

### DSCI 475 - Topological Data Analysis Final Project

```python
#import libraries
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn import manifold, datasets
```

```python
#create Swiss Roll with 3000 samples
X, color = datasets.make_swiss_roll(n_samples=3000, random_state=0)
# Create a colormap object
cmap = plt.cm.inferno
#plot 3D Swiss Roll
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection = "3d")
fig.add_axes(ax)
ax.scatter(
    X[:, 0], X[:, 1], X[:, 2], c = color, cmap=cmap,  s = 10, alpha = 0.8
)
ax.set_title("Original 3D Geometry (Euclidean Space)")
ax.view_init(azim = -66, elev = 12)
_ = ax.text2D(0.8, 0.05, s="n = 3,000", transform = ax.transAxes)
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_3d.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_3d_no_grid.png', transparent=True)
```

```python
# run Isomap on the points in X with 2 dimensional output
Y = manifold.Isomap(n_neighbors = 10, n_components = 2).fit_transform(X)

# 2D projection
fig = plt.figure()
scatter(Y[:,0], Y[:,1], c = color, cmap=cmap, s = 30, alpha = 0.8)
title("Swiss Roll in 2D (X/Y)")
figtext(0.8, 0.05, s = "n = 3,000")
plt.xlabel("Isomap Dimension 1")
plt.ylabel("Isomap Dimension 2")
show()
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_no_grid.png', transparent=True)
```

```python
#Swiss Roll with hole, 3000 samples again
X_hole, color_hole = datasets.make_swiss_roll(n_samples = 3000, hole = True, random_state = 0)

#Plot in 3D
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection = "3d")
fig.add_axes(ax)
ax.scatter(
    X_hole[:, 0], X_hole[:, 1], X_hole[:, 2], c = color_hole, cmap=cmap, s = 10, alpha = 0.8
)
ax.set_title("Swiss Role with Hole in 3D (Euclidean Space)")
ax.view_init(azim = -66, elev = 12)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
_ = ax.text2D(0.8, 0.05, s = "n = 3,000", transform = ax.transAxes)
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_3d_hole.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_3d_hole_no_grid.png', transparent=True)
```

```python
fig = plt.figure()
scatter(X_hole[:, 0], X_hole[:, 1], c = color_hole, cmap=cmap, s = 30, alpha = 0.7)
title("Swiss Role with Hole in 2D (Feature Selection) (X/Y)")
figtext(0.8, 0.05, s = "n = 3,000")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole_no_grid.png', transparent=True)
```

```python
fig = plt.figure()
scatter(X_hole[:, 0], X_hole[:, 2], c = color_hole, cmap=cmap, s = 30, alpha = 0.7)
title("Swiss Role with Hole in 2D (Feature Selection) (X/Z)")
figtext(0.8, 0.05, s = "n = 3,000")
plt.xlabel("X-axis")
plt.ylabel("Z-axis")
plt.show()
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole_2.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole_2_no_grid.png', transparent=True)
```

```python
# run Isomap on the points in X with 2 dimensional output
Y_hole = manifold.Isomap(n_neighbors = 10, n_components = 2).fit_transform(X_hole)

# 2D projection
fig = plt.figure()
scatter(Y_hole[:,0], Y_hole[:,1], c = color_hole, cmap=cmap, s = 30, alpha = 0.7)
title("Swiss Role with Hole in 2D (Feature Extraction) (ISOMAP)")
figtext(0.8, 0.05, s = "n = 3,000")
plt.xlabel("New Component 1")
plt.ylabel("New Component 2")
plt.show()
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole_iso.png', transparent=True)
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
plt.axis('off')
fig.savefig(r'..\Writeup Files/Swiss Roll/swiss_roll_2d_hole_iso_no_grid.png', transparent=True)
```
